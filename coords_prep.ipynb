```python
import pandas as pd 
import numpy as np
```

```python
map_path = "1541875_locations_map.csv"
map = pd.read_csv(map_path, header=None)
map = map.loc[:, :1]
print(map)
```

```python
rng = np.random.default_rng(47293)

selection = rng.choice(map.shape[0], 100000, replace=False)

print(selection)
print(len(selection))
print(len(np.unique(selection)) == len(selection), "unique elements")
```

```python
final_data = map.loc[selection].sort_index()
print(final_data)
```

```python
final_data.to_csv("all_data.csv")
```
